```python
# dependencies
import matplotlib.pyplot as plt # type: ignore
import numpy as np
```

```python
#light-accuracy

x = [100, 250, 450, 700, 1000]
y_alex = [62, 95, 82, 78, 58]
y_vgg = [82, 91, 70, 74, 76]
y_res = [78, 92, 80, 74, 50]

plt.plot(x, y_alex, label= "AlexNet", color='midnightblue')
plt.plot(x, y_vgg, label= "VGGNet", color='purple')
plt.plot(x, y_res, label= "ResNet", color='#82cafc')
plt.grid()
plt.xlabel('Light intensity (LUX)')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.title('Accuracy without noise')
```

```python
#average response time 
alex = [25.9, 28.9, 26.1, 24.6, 26.7, 36.1, 40.8, 41.8]
vgg = [18.7, 23.8, 20.8, 18.5, 20.6, 31.0, 37.7, 36.8]
res = [40.6, 51.6, 51.9, 47.4, 50.4, 95.3, 122.4, 102.2]

alex_avg = sum(alex)/len(alex)
vgg_avg = sum(vgg)/len(vgg)
res_avg = sum(res)/len(res)

moedls = ["AlexNet", "VGGNet", "ResNet"]
means = [alex_avg, vgg_avg, res_avg]

plt.bar(moedls, means, width= 0.5, color="#b790d4")
plt.title('Average elaboration time')
plt.xlabel('Models')
plt.ylabel('time (ms)')
plt.grid()
plt.show()
```

```python
# accuracy drop in terms of light
ba = 95
bv = 91
br = 92

x = [100, 250, 450, 700, 1000]
alex = [62, 95, 82, 78, 58]
vgg = [82, 91, 70, 74, 76]
res = [78, 92, 80, 74, 50]

a_drop = []
v_drop = []
r_drop = []

for elem in alex:
    a_drop.append(ba-elem)

for elem in vgg:
    v_drop.append(bv-elem)

for elem in res:
    r_drop.append(br-elem)

plt.plot(x, a_drop, label= "AlexNet", color='#607c8e')
plt.plot(x, v_drop, label= "VGGNet", color='#06c2ac')
plt.plot(x, r_drop, label= "ResNet", color='#8eab12')
plt.grid()
plt.xlabel('Light intensity (LUX)')
plt.ylabel('Accuracy drop (%)')
plt.legend()
plt.title('Accuracy drop without noise')
```

```python
#accuracy drop in term of noise
ba = 95
bv = 91
br = 92

x = [10, 30, 50]
alex = [86, 62, 62]
vgg = [84, 82, 78]
res = [82, 80, 74]

a_drop = []
v_drop = []
r_drop = []

for elem in alex:
    a_drop.append(ba-elem)

for elem in vgg:
    v_drop.append(bv-elem)

for elem in res:
    r_drop.append(br-elem)


plt.plot(x, a_drop, label= "AlexNet", color='#607c8e')
plt.plot(x, v_drop, label= "VGGNet", color='#06c2ac')
plt.plot(x, r_drop, label= "ResNet", color='#8eab12')
plt.grid()
plt.xlabel('Noise (%)')
plt.ylabel('Accuracy drop (%)')
plt.legend()
plt.title('Accuracy drop @ 250 LUX')
```

```python
#time over light

x = [100, 250, 450, 700, 1000]
t_a = [26.7, 24.6, 26.1, 28.9, 25.9]
t_v = [20.6, 18.5, 20.8, 23.8, 18.7]
t_r = [50.4, 47.4, 51.9, 51.6, 40.6]


plt.plot(x, t_a, label= "AlexNet", color='midnightblue')
plt.plot(x, t_v, label= "VGGNet", color='purple')
plt.plot(x, t_r, label= "ResNet", color='#82cafc')
plt.grid()
plt.xlabel('LUX')
plt.ylabel('ms')
plt.legend()
plt.title('Average Elaboration time')
```
